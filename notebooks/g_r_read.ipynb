```python
def read_gr(path):
    ar = {}
    g_r_spce = open(path,"r").read()
    s_g_r_spce = g_r_spce.split("\n")[4:-1]
    counter = 0
    for linha in s_g_r_spce:
        if counter not in list(ar.keys()):
            ar[counter] = []


        if len(linha.split())==8:
            ar[counter] += [linha.split()]

        if len(linha.split())==2:
            counter+=1

    for key in list(ar.keys()):
        ar[key] = np.array(ar[key]).astype(float)
    return ar
```

```python
gr_spce = read_gr("../spce/g_r.txt")
gr_tip4p = read_gr("../tip4p/g_r.txt")
gr_2005 = read_gr("../2005/g_r.txt")
gr_spcfw = read_gr("../spcfw/g_r.txt")
gr_spcfd = read_gr("../spcfd/g_r.txt")
```

```python
plt.figure(figsize=(10,10))

plt.plot(gr_spce[0][:,1],gr_spce[0][:,2],lw=2)
plt.plot(gr_tip4p[0][:,1],gr_tip4p[0][:,2],lw=2)
plt.plot(gr_2005[0][:,1],gr_2005[0][:,2],lw=2)
plt.plot(gr_spcfw[0][:,1],gr_spcfw[0][:,2],lw=2)
plt.plot(gr_spcfd[0][:,1],gr_spcfd[0][:,2],lw=2)
plt.legend("SPCE TIP4P TIP4P/2005 SPCFW SPCFD".split())
plt.xlim(2,12)
plt.ylim(0)
plt.grid(True)
```
